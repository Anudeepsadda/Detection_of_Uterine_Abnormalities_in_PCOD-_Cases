```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
df=pd.read_csv("/content/pcod data set.csv")
```

```python
# Rename columns for consistency
df.columns = ['Age', 'Current weight', '3 months ago weight', 'Profession',
              'Any sudden weight gain/loss', 'Regular/irregular periods',
              'Menstrual cycle occurrence', 'Facial hair', 'Acne related problems',
              'Exercise/yoga/meditation', 'Sleeping hours',
              'No. of Fruits and vegetables serving/day',
              'No. of junk food servings/day', 'Relaxation activities', 'Target class']
```

```python
# Fill missing values
df.fillna({
    "Age": "None",
    "Current weight": "0.0",
    "Any sudden weight gain/loss": "none",
    "Menstrual cycle occurrence": "none"
}, inplace=True)

# Apply Label Encoding to categorical variables
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
```

```python
# Split dataset into features and target
X = df.drop(['Target class'], axis=1)
Y = df['Target class']
```

```python
# Split into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=0)
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train Decision Tree Classifier (Gini Index)
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=100)
clf_gini.fit(X_train, Y_train)
```

```python
# Predict and evaluate
Y_pred_gini = clf_gini.predict(X_test)
print("Gini Decision Tree Accuracy:", metrics.accuracy_score(Y_test, Y_pred_gini) * 100)
```

```python
# Train Decision Tree Classifier (Entropy)
clf_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=3, min_samples_leaf=5, random_state=100)
clf_entropy.fit(X_train, Y_train)
```

```python
# Predict and evaluate
Y_pred_entropy = clf_entropy.predict(X_test)
print("Entropy Decision Tree Accuracy:", metrics.accuracy_score(Y_test, Y_pred_entropy) * 100)
```
